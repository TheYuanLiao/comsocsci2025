```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("taxis")
```

```python
df["payment"].drop_duplicates()
```

## Groupby operations

```python
total_credit_card = 0 
total_cash = 0 
total_nan = 0
n_credit_card = 0
n_cash = 0
n_nans = 0
for payment_type, price in df[["payment","total"]].values:
    if payment_type=="credit card": 
        total_credit_card+=price
        n_credit_card+=1
    elif payment_type=="cash": 
        total_cash+=price
        n_cash+=1
    else: 
        total_nan+=price
        n_nans+=1
print("average credit card: ", total_credit_card/n_credit_card)
print("average cash: ", total_cash/n_cash)
print("average nan: ", total_nan/n_nans)
```

```python
df["payment"] = df["payment"].fillna("nan")
df.groupby("payment")["total"].mean()
```

## Merge dataframes

```python
df_areas = df[["pickup_zone"]].drop_duplicates()
df_areas["latitude"] = np.random.rand(len(df_areas))
df_areas["longitude"] = np.random.rand(len(df_areas))
```

```python
latitude_dictionary = dict(df_areas[["pickup_zone","latitude"]].values)
df["latitude"] = df["pickup_zone"].map(latitude_dictionary)
```

```python
pd.merge(df, df_areas, on = "pickup_zone")
```
